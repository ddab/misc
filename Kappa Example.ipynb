```python
import matplotlib.pyplot as plt
import random

def twoByTwo():
    a = random.randint(1,100)
    b = random.randint(1,100)
    c = random.randint(1,100)
    d = random.randint(1,100)

    obsAgree = (a+d)/(a+b+c+d)

    aMargin = ((a+b)*(a+c))/(a+b+c+d)
    bMargin = ((c+d)*(b+d))/(a+b+c+d)
    expAgree = (aMargin+bMargin)/(a+b+c+d)

    kappa = (obsAgree-expAgree)/(1-expAgree)
    
    plt.scatter(obsAgree, kappa, s=5)

for i in range(1000):
    twoByTwo()

plt.xlabel('Observed Agreement', fontsize=10)
plt.ylabel('Kappa Score', fontsize=10)
plt.title("Obsevered Agreement vs. Kappa Score")
plt.show()
```
